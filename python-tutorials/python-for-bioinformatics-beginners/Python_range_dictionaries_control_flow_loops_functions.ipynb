## Loops ##

Allows us to repeat actions.\
Two types of loops:\
     -_for_ loops\
     -_while_ loops

### for Loop ###
Used to terate over a sequence such as a list, tuple, string or range.\
Enabling a block of codes to be repeated for each iterm.\
**Syntax**\
 _for placeholder in iterable_:
>    _action_

placeholder - a variable name  \
iterable - any data structure that can be iterated on 

```python
#example 1- print items in a list (or any iterable can be used )
nucleotides = ["A","T","G","C"]
for base in nucleotides:
    print(base)
```

```python
#example 2- print items in a list if item is greater than 10
scores = [10,2,12,15]
for score in scores:
    if score > 10:
        print(score)
```

### while Loop ### 
Enables a block of codes to be repeated so far as a condition (or set of conditions) are satisfied.\
**Syntax**\
_while_ condition:
>     action


```python
#example- add 1 to a variable so far as the variable is less than 5
a = 1 
while a < 5:
     a += 1
     print(a)
```

### Skip current iteration
This is done using the command _continue_.\
This works for both the _for_ and _while_ loops.

```python
sequence="AGNTATNN"
for base in sequence:
    if base == "N":
        continue
    print(base)
```

### Break out of a loop
This is done using the command _break_.\
This works for both the _for_ and _while_ loops.

```python
sequence="AGNTATNN"
for base in sequence:
    if base == "N":
        break
```

### Compute GC Content of  Sequence ###


```python
#Calculating percentage GC of a sequence
seq="AGTTAGCTAGGAG"
gc=seq.count('G')+seq.count('C')
length_of_seq=len(seq)
gc_percentage=gc/length_of_seq * 100
print('percentage gc is:',gc_percentage)

gc_rounded=round(gc_percentage,2)
print('percentage gc rounded:',gc_rounded)

```

### A function to compute GC Content ###

```python
### A function to compute the gc content of sequences

def GC_Content(seq):
    seq="AGTTAGCTAGGAG"
    gc=seq.count('G')+seq.count('C')
    length_of_seq=len(seq)
    gc_percentage=gc/length_of_seq * 100
    gc_rounded=round(gc_percentage,2)
    print('percentage gc:',gc_rounded)

GC_Content('AGGAGAGAT')
GC_Content('AGAAGGAG')
```

### Find the complements of multiple nucleotide sequences
  - Use a
      - loop
      - function
      - dictionary
      - list

```python
#APPROACH: COMBINE FUNCTIONS AND LOOPS
```

```python
DNASEQUENCES=['ATCGA','TTAGC','ACCTAG']
```

```python
len(DNASEQUENCES)
```

```python
# create a function that will find the complement of a nucleotide sequence
# then reuse the function for all your sequences

def sequence_complement_finder(sequence):
    base_complements={'A':'T','T':'A','C':'G','G':'C'}
    sequence_complement=[base_complements[base] for base in sequence]
    sequence_complement=''.join(sequence_complement)
    return sequence_complement
```

```python
# use a 'for' loop to iterate over the list of sequences and find their respective complements

for sequence in DNASEQUENCES:
    sequence_complement=sequence_complement_finder(sequence)
    print('original sequence:  ',sequence)
    print('sequence complement:',sequence_complement)
```

```python
complement_sequences=[]
for sequence in DNASEQUENCES:
    sequence_complement=sequence_complement_finder(sequence)
    complement_sequences.append(sequence_complement)
```

```python
print(complement_sequences)
```

```python
print('original:  ',DNASEQUENCES)
print('complement:',complement_sequences)
```

### Codon Translator ###

```python
sequence = "ATGTTTTAACCC"
codon_table = {
    "ATG": "M",
    "TTT": "F",
    "TAA": "STOP"
}
for i in range(0, len(sequence), 3):
    codon = sequence[i:i+3]
    
    if codon in codon_table:
        if codon_table[codon] == "STOP":
            break
        print(codon_table[codon], end="")
    else:
        print("X", end="")
```

### Detect Invalid bases

```python
sequence = "ATGNCGTAN"

valid_bases = ["A", "T", "G", "C"]

for i, base in enumerate(sequence):
    if base not in valid_bases:
        print("Invalid base at position:", i)
```
