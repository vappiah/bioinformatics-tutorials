# LISTS

-Lists are linear collection of items.\
-Lists are created by putting items inside a square bracket [ ].\
eg. ['A','G','T','C']  or [1,2,3].\
-Individual items are separated by commas.\
-Lists can contain objects of different datatypes including lists.\
eg. [1, 'A', 3.2]   or [[1,2],['A','T']].\
-Items in a list are accessed using indexing.\
-A list is mutable, meaning its content can be changed.\
-New items are added to the tail (end of the list).\
-Iterations can be performed on lists.\
-If all items in a list are strings, then the entire list can be converted 
 to a single string object.

### Accessing items in a list (Indexing and slicing)


-This is done by using indexing and slicing.\
-To access an item or items we specify an **index** or a range of indexes(**start and end**) and in some cases a **step or stride** value.

**Syntax**\
L[m,n,s]\
**-L**:List\
**-m**:start\
**-n**:end\
**-s**:step/stride

```python
codons=['AAA','UUC','CCA','CAC']
```

```python
codons[0]
```

```python
codons[1]
```

```python
len(codons)
```

```python
codons[3]
```

#### Short-hand indexing

 This is done by omitting either the first and or last index.\
 **L[:n]** starts the slice from the beginning of the 
   list to the index just before n.\
 **L[m:]** starts slice from index m to the end of the 
   list.\
 **L[:]** returns a copy of the entire list.

```python
sequences=['AGAGGA','AGAAGGGG','GAGAG','GAGAGAG','AGAGAGA']
```

```python
sequences[:4]
```

```python
sequences[2:]
```

```python
sequences[:]
```

### Iterating over a list

There are different ways to iterate over a list.
In this tutorial we will look at how to use **for** loop and a **list comprehension**

#### for loop

```python
bases=['A','T','G','C']
```

```python
for i in bases:
    print(i)
```

```python
for base in bases:
    print(base)
```

```python
for item in bases:
    print(item)
```

```python
numbers=[1,2,3,4,5,6]
```

```python
for number in numbers:
    print(number)
```

```python
for number in numbers:
    print(number+1)
```

### List comprehensions

This is a shorter syntax for iterating over a list as well as creating new ones. The syntax includes a loop statement.

```python
bases=['A','T','G','C']
```

```python
[base for base in bases]
```

```python
numbers=[1,2,3,4,5,6]
```

```python
[number+1 for number in numbers]
```

```python
result=[number+1 for number in numbers]
```

```python
result
```
