```python
#approach 2
df.drop(["FOLDCHANGE"], axis=1 )
```

```python
#save changes to a new dataframe
newdf=df.drop(["FOLDCHANGE"], axis=1 )
```

```python
#to overwrite the dataframe use 
df.drop(["FOLDCHANGE"], axis=1 ,inplace=True)
```

```python
#remove / drop rows
df.drop(['GeneA'], axis=0,inplace=True)
```

```python
#rename columns
newnames={ "pvalue":"PVALUE"}
df.rename( { "pvalue":"PVALUE"}, axis=1  )
```

```python
#rename rows
new_row_names={ "LBH":"LBH1" }
df.rename(new_row_names, axis=0)
```
